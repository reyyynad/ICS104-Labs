Each line of a text-file scores.txt contains the ID of a student and his grades in 4 exams. Write a python program that reads scores.txt and writes to an output file output.txt the ID , the worst exam grade, the best exam grade and the average grade for each student

```python
# This code create a file

# Do not change Code
# Just Run it
scores = [
'',
'201100020 45.0 50.0 30.0 50.0',
'201100030 97.0 90.0 90.0 95.0',
'201100040 60.0 80.0 75.0 75.0',
'201100050 62.0 60.0 70.0 60.0'
]

try:
    outFile = open("scores.txt","w")

    for s in scores:
        outFile.write(s+"\n")
        
        
except IOERROR:    
    print("Error")
    
finally:
    outFile.close()
```

```python
infile = open("scores.txt", "r")
for line in infile:
    line = line.strip()
    student = line.split()
    scores = student[1:]
    
    for i in range(len(scores)):
        scores[i] = float(scores[i])
        
        
    average = sum(scores) / len(scores)
    mx = max(scores)
    print(scores, mx, average)
infile.close()
        
```
